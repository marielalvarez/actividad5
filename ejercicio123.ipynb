# Ejercicio 1 (Numpy)

```python
import numpy as np
# Crear un vector “x” con 10 ceros y mostrarlo
x = np.zeros(10)
print(x)
```

```python
# Modificar el quinto elemento del vector x para que su valor sea 3.
x[5] = 3
print(x)
```

```python
# Crear un vector “y” con valores del 1 al 20, de 1 en 1. Utiliza la función arange.
y = np.arange(1,21)
print(y)
```

```python
#  Crear un vector “z” con valores del 1 al 20, de 2 en 2.
z = np.arange(1,21,2)
print(z)
```

```python
# Crear una matriz w de 3x2, con valores iniciales en 0.
w = np.zeros((3, 2))
print(w)
```

```python
# Modificar el elemento en la posición 2,2 de la matriz w para que su valor sea 5.
w[1][1] = 5
print(w)
```

```python
#  Crear un vector de 5 elementos, con valores aleatorios enteros entre 1 y 10.
vector_aleatorio = np.random.randint(1, 11, size=5)
print(vector_aleatorio)
```

```python
# Crear el arreglo x1 de la lista [1,2,3]
x1 = np.array([1,2,3])
print(x1)
```

```python
# Sumar 10 a cada elemento del vector
x1 = x1 + 10
print(x1)
```

# Ejercicio 2

```python
# Generar un DataFrame con los datos del archivo.
import pandas as pd
df = pd.read_csv('/titanic.csv')

'''Mostrar por pantalla las dimensiones del DataFrame,
el número de datos que contiene, los nombres de sus columnas y filas,
los tipos de datos de las columnas, las 10 primeras filas y las 10 últimas filas
'''
print('Dimensiones del dataframe: \n', df.shape)
print('Número de datos del dataframe: \n', df.size)
print('Nombre de columnas: \n', df.columns)
print('Nombre de filas: \n', df.index)
print('Tipos de datos columnas: \n', df.dtypes)
print('10 primeras filas: \n', df.head(10))
print('10 ultimas filas: \n', df.tail(10))
```

```python
# Mostrar por pantalla los datos del pasajero con identificador 148.
print(df.loc[df['PassengerId'] == 148])
```

```python
# Mostrar por pantalla las filas pares del DataFrame
print(df.iloc[range(0,len(df),2)])
```

```python
# Mostrar por pantalla los nombres de las personas que iban en primera clase ordenadas alfabéticamente.
print(df[df['Pclass'] == 1]['Name'].sort_values())
```

```python
# Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron.
print(df['Survived'].value_counts(normalize=True) * 100)
```

```python
# Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.
agrupado = df.groupby('Pclass')
print(agrupado['Survived'].value_counts(normalize=True) * 100)
```

```python
# Eliminar del DataFrame los pasajeros con edad desconocida.
print(df.dropna(subset=['Age'])) # subset es la columna que deseas
```

```python
# Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.
mujeres = df[df['Sex'] == 'female']
mujeres_edad = mujeres.groupby('Pclass')
print(mujeres_edad['Age'].mean())
```

```python
# Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no.
df['Menor'] = df['Age'] < 18
print(df)
```

```python
# Mostrar por pantalla el porcentaje de menores y mayores de edad que sobrevivieron en cada clase.
grouped_edad = df.groupby(['Pclass', 'Menor'])
print(grouped_edad['Survived'].value_counts(normalize=True) * 100)
```

#  Ejercicio 3

```python
# Generar un dataframe con los datos del archivo.
db = pd.read_csv('/casasboston.csv')
# Reduce el dataframe utilizando y renombrando las columnas de acuerdo a los siguiente:
db = db.rename(columns = {"TOWN":"CIUDAD",
"CRIM":"INDICE_CRIMEN",
"INDUS":"PCT_ZONA_INDUSTRIAL",
"CHAS":"RIO_CHARLES",
"RM":"N_HABITACIONES_MEDIO",
"MEDV":"VALOR_MEDIANO",
"LSTAT":"PCT_CLASE_BAJA",
"AGE" : "EDAD"})

print(db)
```

```python
# Muestra los datos descriptivos para cada columna (conteo, media, cuartiles)
print(db.describe())
```

```python
# Genera un histograma de número de habitaciones promedio
import matplotlib.pyplot as plt
plt.hist(db['N_HABITACIONES_MEDIO'])
plt.xlabel('Número de habitaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de número de habitaciones')
plt.show()
```

```python
# Genera un diagrama de dispersión que muestre el índice de criminalidad vs la zona
plt.figure(figsize=(10, 5))
plt.scatter(db["PCT_ZONA_INDUSTRIAL"], db["INDICE_CRIMEN"], color="blue", alpha=0.6)
plt.xlabel("Ciudad")
plt.ylabel("Índice de Criminalidad")
plt.title("Índice de Criminalidad vs Ciudad")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Agrupar por ciudad y calcular el valor mediano de la propiedad
valor_median_por_ciudad = db.groupby('CIUDAD')['VALOR_MEDIANO'].median()

plt.figure(figsize=(10, 5))
plt.bar(valor_median_por_ciudad.index, valor_median_por_ciudad.values)

plt.xlabel('Ciudad')
plt.ylabel('Valor Mediano')
plt.title('Valor Mediano de la Propiedad por Ciudad')
plt.xticks(rotation=45)

plt.show()
```

```python
# Realiza un diagrama de caja para la variable de edad
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.boxplot(x=db['EDAD'])

plt.xlabel('Edad')
plt.title('Diagrama de Caja de la Edad')

plt.show()
```
